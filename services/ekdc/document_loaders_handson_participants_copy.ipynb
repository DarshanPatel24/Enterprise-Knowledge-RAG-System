# **Document Loaders**
Document loaders provide a **standard interface** for reading data from different sources (such as Slack, Notion, or Google Drive) into LangChain’s **Document** format. This ensures that data can be handled consistently regardless of the source.

## **Docling Document Loader**
Docling is an Opensource Project by IBM.

Docling parses PDF, DOCX, PPTX, HTML, and other formats into a rich unified representation including document layout, tables etc., making them ready for generative AI workflows like RAG.

## **Document Loaders API Interface**
Each document loader may define its own parameters, but they share a common API:
- **load()** – Loads all documents at once.
- **lazy_load()** – Streams documents lazily, useful for large datasets.

## **Example Usage**
```python
from langchain_docling.loader import DoclingLoader

FILE_PATH = "https://arxiv.org/pdf/2408.09869"

loader = DoclingLoader(file_path=FILE_PATH)

# Load all documents
documents = loader.load()

# For large datasets, lazily load documents
for document in loader.lazy_load():
    print(document)
```

Document loaders provide a "load" method for loading data as documents from a configured source. They optionally implement a "lazy load" as well for lazily loading data into memory.

## **ExportType**
DoclingLoader supports two different export modes:

- **ExportType.DOC_CHUNKS (default):** convert the document into Docling's internal representation, then apply Docling's chunking algorithm, returning one LangChain Document per chunk.
- **ExportType.MARKDOWN:** convert the document into Markdown and return one LangChain Document per input file.

## **Loading Web Page**

````python
# !pip install beautifulsoup4
from langchain_docling.loader import DoclingLoader
from langchain_docling.loader import ExportType

FILE_PATH = "https://docs.langchain.com/oss/python/integrations/providers/all_providers"

loader = DoclingLoader(file_path=FILE_PATH, export_type=ExportType.DOC_CHUNKS)

# Load all documents
documents = loader.load()
````
